# Combined HVAC + Battery Thermal & Energy Impact (Concept Model)

This notebook combines:
- Traction power demand (simple speed-based model)
- HVAC electrical load (user-defined scenarios)
- Battery heat generation (Joule heating): Q = IÂ²R, I = P/V
- Lumped thermal model: dT/dt = (Q - UA*(T - Tamb) - Q_cool)/Cth

Goal: show how HVAC and thermal management affect energy consumption and battery temperature trends.
All parameters are illustrative and intended for concept trade studies.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Parameters (illustrative)

You can tune these to reflect different vehicle classes (compact vs SUV, 400V vs 800V) while staying non-proprietary.

```python
# --- Simulation settings ---
dt = 1.0  # s
duration_min = 40
t = np.arange(0, duration_min*60 + dt, dt)

# --- Environment ---
T_amb = 0.0     # degC (cold day to make HVAC meaningful)
T0 = 15.0       # degC initial pack temp

# --- Battery electrical ---
V_pack = 400.0  # V
R_int  = 0.050  # ohm (illustrative equivalent pack resistance)

# --- Battery thermal ---
m_batt = 450.0      # kg
cp_batt = 1000.0    # J/kg-K
Cth = m_batt * cp_batt
UA = 80.0           # W/K heat rejection to ambient/underbody/coolant loop (effective)

# --- Battery cooling control (simple thermostat) ---
T_set = 35.0         # degC start cooling above this
cooling_max_w = 3500 # W max heat removal capability (illustrative)
cooling_gain = 350.0 # W/K above setpoint (ramp rate)

# --- Cooling electrical penalty (very simplified) ---
# Assume COP ~ 2.5 for cooling loop/compressor equivalent (illustrative)
COP_cool = 2.5
```

```python
def traction_power_kw(speed_mps, mass_kg=2200.0, Crr=0.010, CdA=0.70, rho=1.2, eff=0.90):
    """
    Very simplified road load power (kW): rolling + aero at steady speed.
    P = (F_roll + F_aero) * v / eff
    """
    g = 9.81
    F_roll = Crr * mass_kg * g
    F_aero = 0.5 * rho * CdA * speed_mps**2
    P_w = (F_roll + F_aero) * speed_mps / max(eff, 1e-6)
    return P_w / 1000.0

def hvac_power_kw_profile(t_seconds, mode="cold_high"):
    """
    Simple HVAC load shapes (kW electrical). Replace with your own.
    """
    if mode == "cold_high":
        # high at start, then taper (cabin warm-up)
        return 4.5*np.exp(-t_seconds/600) + 1.5
    if mode == "mild":
        return 1.2*np.ones_like(t_seconds)
    if mode == "hot_ac":
        return 3.0*np.exp(-t_seconds/900) + 1.0
    return 0.0*np.ones_like(t_seconds)

# Scenario: constant speed cruise
speed_kph = 90
speed_mps = speed_kph / 3.6
P_trac = traction_power_kw(speed_mps) * np.ones_like(t)  # kW

# Choose HVAC scenario
hvac_mode = "cold_high"
P_hvac = hvac_power_kw_profile(t, mode=hvac_mode)  # kW
```

```python
def joule_heat_w(P_total_kw, V=V_pack, R=R_int):
    """
    Q = I^2 R, I = P/V
    P_total_kw includes traction + HVAC + cooling electrical
    """
    P_w = P_total_kw * 1000.0
    I = P_w / V
    return (I**2) * R

def cooling_heat_removed_w(T, Tset=T_set, gain=cooling_gain, Qmax=cooling_max_w):
    """
    Simple thermostat: remove heat only above setpoint.
    """
    if T <= Tset:
        return 0.0
    return min(gain * (T - Tset), Qmax)

def simulate_combined(P_trac_kw, P_hvac_kw, t_seconds):
    n = len(t_seconds)
    T = np.zeros(n); T[0] = T0
    Q_gen = np.zeros(n)
    Q_cool = np.zeros(n)
    P_cool_elec = np.zeros(n)
    P_total = np.zeros(n)

    for i in range(1, n):
        # cooling based on previous temp
        Q_cool[i] = cooling_heat_removed_w(T[i-1])
        P_cool_elec[i] = (Q_cool[i] / COP_cool) / 1000.0  # kW

        P_total[i] = P_trac_kw[i] + P_hvac_kw[i] + P_cool_elec[i]
        Q_gen[i] = joule_heat_w(P_total[i])

        dTdt = (Q_gen[i] - UA*(T[i-1] - T_amb) - Q_cool[i]) / Cth
        T[i] = T[i-1] + dTdt * dt

    return T, P_total, Q_gen, Q_cool, P_cool_elec

T, P_total, Q_gen, Q_cool, P_cool_elec = simulate_combined(P_trac, P_hvac, t)
```

```python
E_trac_kwh = np.sum(P_trac) * dt / 3600.0
E_hvac_kwh = np.sum(P_hvac) * dt / 3600.0
E_cool_kwh = np.sum(P_cool_elec) * dt / 3600.0
E_total_kwh = np.sum(P_total) * dt / 3600.0

summary = pd.DataFrame([{
    "speed_kph": speed_kph,
    "hvac_mode": hvac_mode,
    "duration_min": duration_min,
    "E_trac_kwh": E_trac_kwh,
    "E_hvac_kwh": E_hvac_kwh,
    "E_batt_cooling_kwh": E_cool_kwh,
    "E_total_kwh": E_total_kwh,
    "T_start_C": float(T[0]),
    "T_end_C": float(T[-1]),
    "T_peak_C": float(np.max(T))
}])

summary
```

```python
modes = ["mild", "cold_high", "hot_ac"]
rows = []

for mode in modes:
    P_h = hvac_power_kw_profile(t, mode=mode)
    Tm, Ptot, Qg, Qc, Pc = simulate_combined(P_trac, P_h, t)
    rows.append({
        "hvac_mode": mode,
        "E_total_kwh": float(np.sum(Ptot)*dt/3600.0),
        "E_hvac_kwh": float(np.sum(P_h)*dt/3600.0),
        "E_cooling_kwh": float(np.sum(Pc)*dt/3600.0),
        "T_end_C": float(Tm[-1]),
        "T_peak_C": float(np.max(Tm))
    })

trade = pd.DataFrame(rows).sort_values("E_total_kwh")
trade
```

## Interpretation (Engineering Notes)

- HVAC is a **time-based** load: it can noticeably raise total energy even if traction demand is unchanged.
- Battery temperature is driven by both:
  - internal heat generation (related to total electrical power), and
  - heat rejection capability (UA) plus any active cooling.
- Active cooling improves thermal margins but adds **electrical energy penalty** (COP-based in this simplified model).

### Portfolio-safe next steps
- Add ambient temperature dependence for HVAC power (simple linear or piecewise)
- Use a simple drive cycle speed trace instead of constant speed

```python
def speed_profile_highway(t_seconds, speed_kph=100):
    return np.full_like(t_seconds, speed_kph/3.6, dtype=float)

def speed_profile_city(t_seconds):
    """
    Synthetic stop-and-go city profile (m/s).
    Repeats every 120 s:
      0â20s: accelerate to 50 kph
      20â60s: cruise
      60â75s: decel to stop
      75â120s: idle / crawl
    """
    t = t_seconds
    v = np.zeros_like(t, dtype=float)
    period = 120

    for i, ti in enumerate(t):
        tau = ti % period
        if tau < 20:
            v[i] = (tau/20) * (50/3.6)
        elif tau < 60:
            v[i] = 50/3.6
        elif tau < 75:
            v[i] = (1 - (tau-60)/15) * (50/3.6)
        else:
            v[i] = (8/3.6) if (tau % 10) < 3 else 0.0
    return v
```

```python
def traction_power_kw_series(speed_mps_series, mass_kg=2200.0, Crr=0.010, CdA=0.70, rho=1.2, eff=0.90):
    g = 9.81
    v = np.array(speed_mps_series, dtype=float)
    F_roll = Crr * mass_kg * g
    F_aero = 0.5 * rho * CdA * v**2
    P_w = (F_roll + F_aero) * v / max(eff, 1e-6)
    return P_w / 1000.0

# HVAC scenario (reuse your earlier function)
hvac_mode = "cold_high"
P_hvac = hvac_power_kw_profile(t, mode=hvac_mode)

# Speed profiles
v_hw = speed_profile_highway(t, speed_kph=100)
v_city = speed_profile_city(t)

# Traction power
P_trac_hw = traction_power_kw_series(v_hw)
P_trac_city = traction_power_kw_series(v_city)
```

```python
T_hw, Ptot_hw, Qg_hw, Qc_hw, Pc_hw = simulate_combined(P_trac_hw, P_hvac, t)
T_city, Ptot_city, Qg_city, Qc_city, Pc_city = simulate_combined(P_trac_city, P_hvac, t)
```

```python
def kwh_from_profile(P_kw):
    return float(np.sum(P_kw) * dt / 3600.0)

def km_traveled(speed_mps_series):
    return float(np.sum(speed_mps_series) * dt / 1000.0)

rows = []

for name, v, P_trac, Tm, Ptot, Pc in [
    ("Highway", v_hw, P_trac_hw, T_hw, Ptot_hw, Pc_hw),
    ("City", v_city, P_trac_city, T_city, Ptot_city, Pc_city),
]:
    dist_km = km_traveled(v)
    E_total = kwh_from_profile(Ptot)

    rows.append({
        "case": name,
        "hvac_mode": hvac_mode,
        "distance_km": dist_km,
        "E_total_kwh": E_total,
        "E_trac_kwh": kwh_from_profile(P_trac),
        "E_hvac_kwh": kwh_from_profile(P_hvac),
        "E_batt_cooling_kwh": kwh_from_profile(Pc),
        "effective_kWh_per_100km": (E_total / max(dist_km,1e-6))*100.0,
        "T_peak_C": float(np.max(Tm))
    })

summary_df = pd.DataFrame(rows)
summary_df
```
